```python
# CRIANDO OBJETOS REFERENTE AO MUNICÍPIO 
class Municipio:
    def __init__(self,id,nome,risco,custo,populacao):
        self.id = id
        self.nome = nome
        self.risco = risco
        self.custo = custo
        self.populacao = populacao
    
    def __str__(self):
        return(
            f"ID: {self.id} | "
            f"Município: {self.nome} | "
            f"Risco: {self.risco}"
        )
```

```python
porto_alegre = Municipio(
    1,
    "Porto Alegre",
    0.72,
    1850,
    1400000
)

print(porto_alegre)
```

```python
class Grafo:

    def __init__(self):
        self.vertices = {}
        self.adjacencias = {}

    def adicionar_municipio(self, municipio):
        self.vertices[municipio.id] = municipio
        if municipio.id not in self.adjacencias:
            self.adjacencias[municipio.id] = []
    
    def adiconar_rota(self, origem, destino, peso):
        self.adjacencias[origem].append(
            (destino, peso)
        )
        self.adjacencias[destino].append(
            (origem, peso)
        )

            
```

```python
pa = Municipio(
    1,
    "Porto Alegre",
    0.72,
    1850,
    1400000
)

canoas = Municipio(
    2,
    "Canoas",
    0.60,
    900,
    350000
)

pelotas = Municipio(
    3,
    "Pelotas",
    0.55,
    700,
    300000
)
```

```python
# MONTANDO O GRAFO

grafo = Grafo()

grafo.adicionar_municipio(pa)
grafo.adicionar_municipio(canoas)
grafo.adicionar_municipio(pelotas)

grafo.adiconar_rota(1, 2, 2.5)
grafo.adiconar_rota(1, 3, 5.0)
grafo.adiconar_rota(2, 3, 3.0)
```

```python
class Node:
    def __init__(self, municipio):
        self.municipio = municipio
        self.left = None
        self.right = None
```

```python
class buscaArvoreBinaria:
    def __init__(self):
        self.root = None # RAIZ DA ÁRVORE COMEÇA VAZIA

    def insert(self, municipio):
        novo_no = Node(municipio)
        if self.root is None:
            self.root = novo_no
        else:
            self._insert_recursivo(
                self.root,
                novo_no
            )
    
    def _insert_recursivo( self, atual, novo_no):
        risco_novo = novo_no.municipio.risco
        risco_atual = atual.municipio.risco

        if risco_novo < risco_atual:
            if atual.left is None:
                atual.left = novo_no
            else:
                self._insert_recursivo(
                    atual.left,
                    novo_no
                )
        else:
            if atual.right is None:
                atual.right = novo_no
            else:
                self._insert_recursivo(
                    atual.right,
                    novo_no
                )

    def in_order(self):
        self._in_order_recursivo(
            self.root
        )

    def _in_order_recursivo(self, node):
        if node is not None:
            self._in_order_recursivo(
                node.left
            )

            print(node.municipio.nome,"-", node.municipio.risco)
            self._in_order_recursivo(
                node.right
            )

    def buscar(self, risco_min, risco_max):
        resultados = []

        self._buscar_recursivo(
            self.root,
            risco_min,
            risco_max,
            resultados
        )

        return resultados
    
    def _buscar_recursivo(self, node, risco_min, risco_max, resultados):
        if node is None:
            return
        
        risco = node.municipio.risco

        if risco_min <= risco <= risco_max:
            resultados.append(
                node.municipio
            )

        self._buscar_recursivo(
            node.left,
            risco_min,
            risco_max,
            resultados
        )

        self._buscar_recursivo(
            node.right,
            risco_min,
            risco_max,
            resultados
        )

    def altura(self):
        return self._altura_recursiva(
            self.root
        )
    
    def _altura_recursiva(self, node):
        if node is None:
            return 0
        
        altura_esquerda = self._altura_recursiva(
            node.left
        )

        altura_direita = self._altura_recursiva(
            node.right
        )

        return max(
            altura_esquerda,
            altura_direita
        ) + 1
    
    def altura(self):
        return self._altura_recursiva(self.root)
    
    def _altura_recursiva(self, node):
        if node is None:
            return 0
        
        altura_esquerda = self._altura_recursiva(node.left)
        altura_direita = self._altura_recursiva(node.right)
        
        return max(
            altura_esquerda,
            altura_direita
        ) + 1
    



```
